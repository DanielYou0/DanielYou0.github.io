```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
from sklearn.svm import SVR
```

Link to the data set : http://lib.stat.cmu.edu/datasets/boston

```python
df = pd.read_csv("housing.data", delim_whitespace=True, header=None)
# Define the features and labels
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Scale the features 
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Train a Linear Regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Train a K-Nearest Neighbors model
knn_regressor = KNeighborsRegressor(n_neighbors=5)
knn_regressor.fit(X_train, y_train)
```

```python
# Train a Support Vector Machine model
svm_regressor = SVR(kernel='rbf')
svm_regressor.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = regressor.predict(X_test)
y_knn_pred = knn_regressor.predict(X_test)
y_svm_pred = svm_regressor.predict(X_test)
```

```python
# Evaluate the performance of the models 
mse = mean_squared_error(y_test, y_pred)
knn_mse = mean_squared_error(y_test, y_knn_pred)
svm_mse = mean_squared_error(y_test, y_svm_pred)

mae = mean_absolute_error(y_test, y_pred)
knn_mae = mean_absolute_error(y_test, y_knn_pred)
svm_mae = mean_absolute_error(y_test, y_svm_pred)

rmse = np.sqrt(mse)
knn_rmse = np.sqrt(knn_mse)
svm_rmse = np.sqrt(svm_mse)

r2 = r2_score(y_test, y_pred)
knn_r2 = r2_score(y_test, y_knn_pred)
svm_r2 = r2_score(y_test, y_svm_pred)

print("Linear Regression Mean Squared Error:", mse)
print("K-Nearest Neighbors Mean Squared Error:", knn_mse)
print("Support Vector Machine Mean Squared Error:", svm_mse)

print("\n")

print("Linear Regression Mean Absolute Error:", mae)
print("K-Nearest Neighbors Mean Absolute Error:", knn_mae)
print("Support Vector Machine Mean Absolute Error:", svm_mae)

print("\n")

print("Linear Regression Root Mean Squared Error:", rmse)
print("K-Nearest Neighbors Root Mean Squared Error:", knn_rmse)
print("Support Vector Machine Root Mean Squared Error:", svm_rmse)

print("\n")

print("Linear Regression R2 Score:", r2)
print("K-Nearest Neighbors R2 Score:", knn_r2)
print("Support Vector Machine R2 Score:", svm_r2)
```

```python
# Plot the actual vs predicted values for Linear Regression
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Prices (Linear Regression)")
plt.show()

# Plot the actual vs predicted values for K-Nearest Neighbors
plt.scatter(y_test, y_knn_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Prices (K-Nearest Neighbors)")
plt.show()

# Plot the actual vs predicted values for Support Vector Machine
plt.scatter(y_test, y_svm_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Prices (Support Vector Machine)")
plt.show()
```
